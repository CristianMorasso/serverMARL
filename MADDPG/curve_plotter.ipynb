```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
import glob
env_name = "simple_speaker_listener_v4"
csv_list = glob.glob(f"seeds_test/{env_name}*.csv")
csv_list
```

```python
res_list = []
# res_list.append(pd.read_csv("simple_speaker_listener_v4_Base.csv"))
# res_list.append(pd.read_csv("simple_speaker_listener_v4_lr0_0001_5k_ep.csv"))
# res_list.append(pd.read_csv("simple_speaker_listener_v4_lr0_001_5k_ep_ca_32.csv"))
# res_list.append(pd.read_csv("simple_speaker_listener_v4_lr0_0001_4k_ep_ca_32.csv"))
# res_list.append(pd.read_csv("simple_speaker_listener_v4_lr0_001_4k_ep_c_64_a_32.csv"))
# res_list.append(pd.read_csv("simple_speaker_listener_v4_lr0_001_4k_ep_c_128_a_32.csv"))
res_list = [pd.read_csv(f) for f in csv_list]
```

```python
for df in res_list:
    df.columns=["episode", "reward"]
```

```python
for i, df in enumerate(res_list):
    # plt.tight_layout(False)
    sns.lineplot(data=df[:50000], x="episode", y = "reward", label=csv_list[i])
    plt.title("Hyper params test")
# sns.legend()
```

```python
df_tot = pd.concat(res_list)
# df_tot
```

```python
sns.lineplot(data=df_tot, x="episode", y = "reward")
# plt.title("Hyper params test")#
```
